```python
# prompt: let dependent variable y be New Hourly Wage and independent variable x as Big Mac Price

import statsmodels.api as sm

# Define dependent and independent variables
y = data['New Hourly Wage']
x = data['Big Mac Price']


```

```python
x
```

```python
# prompt: extract simple linear regression on x and y

# Assuming 'x' and 'y' are already defined as in your previous code

# Add a constant to the independent variable
x = sm.add_constant(x)

# Fit the linear regression model
model = sm.OLS(y, x).fit()

# Print the model summary
print(model.summary())
```

```python
# prompt: Check if residual is normally distributed

import matplotlib.pyplot as plt
# Assuming 'model' is the fitted OLS model from your previous code

import scipy.stats as stats

# Get the residuals
residuals = model.resid

# Create a histogram of the residuals
plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=30, edgecolor='black')
plt.title('Histogram of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()

# Create a Q-Q plot of the residuals
plt.figure(figsize=(8, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of Residuals')
plt.show()

```

```python
# prompt: Check if residual is normally distributed, plot histogram with a line

# Assuming 'model' is the fitted OLS model and 'residuals' are already calculated

import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

# Create a histogram of the residuals with a fitted normal distribution line
plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=30, density=True, alpha=0.6, color='g', label='Residuals') #density=True for probability density

# Fit a normal distribution to the residuals
mean_res = np.mean(residuals)
std_res = np.std(residuals)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)  # Generate x values for the normal distribution line
p = stats.norm.pdf(x, mean_res, std_res)  # Calculate probability density function values
plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution') #plot normal distribution with k color


plt.title('Histogram of Residuals with Fitted Normal Distribution')
plt.xlabel('Residuals')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# prompt: conduct white test to see whether linear regression is homoscadiscity or not

# Assuming 'model' and 'data' are already defined as in your previous code

import statsmodels.stats.api as sms

# Perform White's test for heteroscedasticity
white_test = sms.het_white(model.resid, model.model.exog)

# Print the results
labels = ['Test Statistic', 'P-value', 'df']
print(dict(zip(labels, white_test)))

# Interpret the results
alpha = 0.05  # Significance level

if white_test[1] < alpha:
    print("Reject the null hypothesis. Heteroscedasticity is present.")
else:
    print("Fail to reject the null hypothesis. No evidence of heteroscedasticity.")
```
