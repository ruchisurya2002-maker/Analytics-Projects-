```python
# prompt: in the engine column take int before cc and in max power and max torque column take only int number before @

import pandas as pd
import re

# Function to extract integer part before 'cc'
def extract_engine_int(engine_str):
    match = re.search(r'(\d+)', engine_str)
    if match:
        return int(match.group(1))
    return None

# Apply the function to the 'Engine' column
data['Engine_int'] = data['Engine'].apply(extract_engine_int)


# Function to extract integer part before '@'
def extract_power_torque_int(value):
  if pd.isna(value):
      return None
  match = re.search(r'(\d+)', value)
  if match:
    return int(match.group(1))
  return None

# Apply the function to 'Max Power' and 'Max Torque' columns
data['Max_power_int'] = data['Max Power'].apply(extract_power_torque_int)
data['Max_torque_int'] = data['Max Torque'].apply(extract_power_torque_int)
```

```python
# prompt: treat the outliers using IQR

import pandas as pd

def treat_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[col] = data[col].clip(lower=lower_bound, upper=upper_bound)
    return data

# Example usage (replace 'Price' with the actual column name):
for col in ['Age', 'Kilometer', 'Length', 'Width', 'Height', 'Price', 'Engine_int', 'Max_power_int', 'Max_torque_int']:
    data = treat_outliers_iqr(data, col)
```

```python
# prompt: define independent x and dependent variable y

# Assuming 'data' DataFrame is already created and processed as in the previous code

x = data.drop('Price', axis=1)
y = data['Price']
```

```python
# prompt: run multiple linear regression using stats model

import statsmodels.api as sm

# Add a constant to the independent variables
x = sm.add_constant(x)

# Fit the multiple linear regression model
model = sm.OLS(y, x).fit()

# Print the model summary
print(model.summary())
```

```python
# prompt: drop insignficant variable one by one on the basis of  p value

import statsmodels.api as sm

def drop_insignificant_variables(x, y):
    # Add a constant to the independent variables
    x = sm.add_constant(x)

    # Initialize variables to store the best model and its p-values
    best_model = None
    best_pvalues = None

    while True:
        # Fit the multiple linear regression model
        model = sm.OLS(y, x).fit()

        # Get the p-values for each variable
        pvalues = model.pvalues

        # Find the variable with the highest p-value
        max_pvalue = pvalues.max()

        # Check if all p-values are below the significance level (e.g., 0.05)
        if max_pvalue < 0.05:
            best_model = model
            best_pvalues = pvalues
            break

        # Identify the variable with the highest p-value
        variable_to_drop = pvalues.idxmax()

        # Remove the variable with the highest p-value
        if variable_to_drop != 'const':
            print(f"Dropping variable '{variable_to_drop}' with p-value {max_pvalue:.3f}")
            x = x.drop(variable_to_drop, axis=1)
        else:
            # If the constant has the highest p-value, there is an issue with the model specification
            #  Handle accordingly (e.g., check for multicollinearity, transformations)
            print("Constant has highest p-value. Consider model respecification.")
            best_model = model
            best_pvalues = pvalues
            break

    return best_model, best_pvalues

# Example usage
x = data.drop('Price', axis=1)
y = data['Price']

best_model, best_pvalues = drop_insignificant_variables(x, y)

print("\nBest Model Summary:")
print(best_model.summary())
```

```python
# prompt: check if residual is normally distributed using histogram using line

import matplotlib.pyplot as plt
import numpy as np
# Assuming 'best_model' from the previous code is the fitted OLS model
residuals = best_model.resid

# Plotting the histogram of residuals with a fitted normal distribution line
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, stat='density')  # Use 'density' for probability density

# Generate a normal distribution line for comparison
from scipy.stats import norm
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, residuals.mean(), residuals.std())
plt.plot(x, p, 'k', linewidth=2)

plt.title('Histogram of Residuals with Normal Distribution Overlay')
plt.xlabel('Residuals')
plt.ylabel('Density')
plt.show()
```

```python
# prompt: generate histogram for homoscadasticity

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'best_model' from the previous code is the fitted OLS model
residuals = best_model.resid

# Create a scatter plot of predicted values vs. residuals
plt.figure(figsize=(8, 6))
plt.scatter(best_model.fittedvalues, residuals)
plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at y=0
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```

```python
# prompt: conduct white test to check homoscadticity of the model and check the hypothesis

import statsmodels.stats.api as sms
# Assuming 'best_model' from the previous code is the fitted OLS model
# Perform the White test for heteroscedasticity
white_test = sms.het_white(best_model.resid, best_model.model.exog)

# Print the test results
labels = ['Test Statistic', 'P-value', 'F-Statistic', 'F-Test P-value']
print(dict(zip(labels, white_test)))

# Interpret the results
alpha = 0.05  # Significance level
if white_test[1] < alpha:
    print("Reject the null hypothesis. Heteroscedasticity is present.")
else:
    print("Fail to reject the null hypothesis. Homoscedasticity is present.")
```
